```python
# @markdown <font color=Red size=4><b>🎬 YouTube downloader</b></font>
# @markdown <center><img src="https://raw.githubusercontent.com/SudoR2spr/SudoR2spr/main/assets/angel-op/Yt-dlp-angel.png" alt="yt-dlp logo" height="90"></center>
# @markdown <p>⬤ Use <b>YouTube Playlist and single link</b> Direct will download your Google drive, Folder Name is <b>YouTube video</b></p>
#URL = "" # @param {type:"string"}
#video_url = https://youtu.be/lyZ4EPre81s?si=fdp_uShL5RSP2aO7 # @param {type:"string"})
OUTPUT_PATH = "/content/drive/My Drive/Youtube video" # @param {type:"string"}

import yt_dlp
import os


ydl_opts = {
    'outtmpl': os.path.join(OUTPUT_PATH, '%(uploader)s/%(title)s.%(ext)s'),
    '-f': 'best',
    'parse-metadata': "%(series)s S%(season_number)02dE%(episode_number)02d:%(title)s",
    'vcodec': 'h264',
    'replace-in-metadata': "title,uploader" "[ angel_]" "-",
}


def download_video(url):
    with yt_dlp.YoutubeDL(ydl_opts) as ydl:
        ydl.download([url])

print(".")

banner = '''
█░█░█ █▀█ █▀█ █▀▄ █▀▀ █▀█ ▄▀█ █▀▀ ▀█▀
▀▄▀▄▀ █▄█ █▄█ █▄▀ █▄▄ █▀▄ █▀█ █▀░ ░█░'''

print(banner)


if __name__ == '__main__':
    video_url = input("🎬 𝐄𝐧𝐭𝐞𝐫 𝐓𝐡𝐞 𝐘𝐨𝐮𝐭𝐮𝐛𝐞 𝐕𝐢𝐝𝐞𝐨 𝐋𝐢𝐧𝐤 > 🔗 ")
    download_video (video_url)
    print(".")
    print("✅ 𝐒𝐮𝐜𝐜𝐞𝐬𝐬𝐟𝐮𝐥𝐥𝐲 𝐃𝐨𝐧𝐞..." "https://drive.google.com/drive/my-drive")
```
